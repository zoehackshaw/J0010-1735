```python
import numpy as np
from astropy.io import fits
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
data_file = open('ast6725_pulsar.dat','rb') # update this with your file location
x = np.fromfile(data_file,dtype = 'float32')
```

```python
print(len(x))
data_file.close()
```

```python
plt.plot(x[:1000])
```

```python
a = np.fft.fft(x) # the Fourier transform; a = complex array
```

```python
p = np.abs(a[:2**18]**2) # power
p[0] = np.abs(a[0]) # first element is different
p[1:] = p[1:] /p[0] # Poisson normalization
```

```python
plt.plot(p[10:2000]) # not much below
```

```python
plt.plot(p[10:20000]) #hmmm ... looks like a pulsar here!
```

```python
#finding frequency bin with peak power
plt.plot(p[15640:15655])
```

```python
k=15640+8 #determined from plot above
print("the frequency bin with the peak power is",k)
```

```python
plt.plot(p[10:50000])
plt.xlabel("Frequency Bins (K)")
plt.ylabel("Power")
```

```python
N=2**19 #from HW question
dt=0.001 #in ms
t=N*dt
print("time is",t,"seconds")
```

```python
fk=k/t
print("the frequency of the pulsar is",fk,"hertz")
```

```python
period=1/fk
print("the period of the pulsar is",period,"seconds")
```

```python
print("Based on the",period,"second period of the pulsar, this is most likely the Crab pulsar")
```
