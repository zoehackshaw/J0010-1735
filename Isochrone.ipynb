```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from statistics import *
from matplotlib.lines import Line2D 
from matplotlib.figure import *
from math import log
```

```python
logt_s=3.717
logg_s=2.58

t_s=10**3.717
```

```python
data = np.genfromtxt("isochrone_data.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
iso11=np.genfromtxt("iso11.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
iso10=np.genfromtxt("iso10.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
iso9=np.genfromtxt("iso9.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
iso14=np.genfromtxt("is014b.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
iso14a=np.genfromtxt("iso14.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
```

```python
#not inverted
plt.plot(data['logt'], data['logg'])
#plt.plot(iso11['logt'], iso11['logg'])
plt.plot(iso10['logt'], iso10['logg'])
plt.scatter(logt_s,logg_s, color='red',marker='s')
plt.xlabel("logTeff")
plt.ylabel("logg")
```

```python
#inverted
plt.figure(dpi=125)
plt.scatter(logt_s,logg_s, color='red',marker='s',label='J0010-1735',zorder=5)
plt.plot(data['logt'], data['logg'],linewidth=4,label='12 Gyrs',zorder=1)
plt.plot(iso11['logt'], iso11['logg'],label='11 Gyrs',zorder=2)
plt.plot(iso10['logt'], iso10['logg'],label='10 Gyrs',zorder=3)
plt.plot(iso9['logt'], iso9['logg'],color='black',label='9 Gyrs',zorder=4)
plt.plot(iso14['logt'], iso14['logg'],color='black',label='14 Gyrs',zorder=5)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel(r"$\log$T$_{eff}$")
plt.ylabel(r"$\log$G")
plt.legend(bbox_to_anchor=(0.7, 0.95), loc='upper left', borderaxespad=0.)
```

```python
#varying alpha
alpha_02=np.genfromtxt("alpha_0.2iso.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
alpha0=np.genfromtxt("alpha0iso.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
alpha02=np.genfromtxt("alpha0.2iso.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
alpha04=np.genfromtxt("alpha0.4iso.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
alpha06=np.genfromtxt("alpha0.6iso.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])
alpha08=np.genfromtxt("alpha0.8iso.txt", delimiter=",", names=["x","a", "b", "logt","logg","c","d","e","f"])

temps_02=[]
for i in alpha_02['logt']:
    t=10**i
    temps_02.append(t)

temps0=[]
for i in alpha0['logt']:
    t=10**i
    temps0.append(t)
    
temps02=[]
for i in alpha02['logt']:
    t=10**i
    temps02.append(t)

temps04=[]
for i in alpha04['logt']:
    t=10**i
    temps04.append(t)

temps06=[]
for i in alpha06['logt']:
    t=10**i
    temps06.append(t)
    
temps08=[]
for i in alpha08['logt']:
    t=10**i
    temps08.append(t)

plt.figure(dpi=125)
plt.scatter(t_s,logg_s,color='red',marker='s',label='J0010-1735',zorder=5)
plt.plot(temps_02, alpha_02['logg'],color='black',label='alpha= -0.2')
plt.plot(temps0, alpha0['logg'],color='blue',label='alpha= 0')
plt.plot(temps02, alpha02['logg'],color='orange',label='alpha= 0.2')
plt.plot(temps04, alpha04['logg'],color='purple',label='alpha= 0.4')
plt.plot(temps06, alpha06['logg'],color='lime',label='alpha= 0.6')
plt.plot(temps08, alpha08['logg'],color='grey',label='alpha= 0.8')
plt.errorbar(t_s,logg_s,yerr=0.2,color='red',zorder=5)

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
#plt.xlim(5400,5000)
#plt.ylim(2,3.2)
plt.xlabel(r"T$_{eff}$ (K)")
plt.ylabel(r"$\log g$")
plt.legend(bbox_to_anchor=(0.7, 0.95), loc='upper left', borderaxespad=0.)
```
