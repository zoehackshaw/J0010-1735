```python
AgePR1=[3.6426,11.25,0.88,10.46,4.99,8.27,20.17,19.38,18.39,14.89,14.85,17.97,3.96,16.58]
AgePR2=[21.574,14.75,14.33,0.2,9.6,-3.68,12.46,31.42,21.34,9.9,19,16.48,25.63,14.85,32.5]
Z1=[57,58,59,60,62,63,64,65,66,68,69,72,76,77]
Z2=[56,57,58,59,60,62,63,64,66,67,68,69,72,76,77]
plt.scatter(Z1,AgePR1)
plt.xlabel("Atomic Number")
plt.ylabel("Age (Gyr)")
plt.title("Age via Th Equation (i) vs Atomic Number")
plt.text(67.5,2.5,mean(AgePR1))
plt.text(67.5,1,stdev(AgePR1))
plt.text(65,2.5,"mean=")
plt.text(65,1,"stdev=")
z=np.polyfit(Z1,AgePR1,1)
p=np.poly1d(z)
plt.plot(Z1,p(Z1),"-")
scipy.stats.linregress(Z1,AgePR1)
```

```python
plt.scatter(Z2,AgePR2)
plt.xlabel("Atomic Number")
plt.ylabel("Age (Gyr)")
plt.title("Age via Th Equation (ii) vs Atomic Number")
plt.text(67.5,2.5,mean(AgePR2))
plt.text(67.5,0,stdev(AgePR2))
plt.text(65,2.5,"mean=")
plt.text(65,0,"stdev=")
z=np.polyfit(Z2,AgePR2,1)
p=np.poly1d(z)
plt.plot(Z2,p(Z2),"-")
scipy.stats.linregress(Z2,AgePR2)
```
